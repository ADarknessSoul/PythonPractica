```python
import pandas as pd
import re

text = """mi chat"""

pattern = r'(\d{2}/\d{2}/\d{2}) (\d{2}:\d{2}) - (.*?): (.*)'
matches = re.findall(pattern, text)

print(matches)

df = pd.DataFrame(matches, columns=['Date', 'Time', 'User', 'Message'])

```
